```python
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(x)
    except:
        return None

df['total_sqft'] = df['total_sqft'].apply(convertRange)
```

```python
df['price_per_sqft'] = df['price'] * 100000 / df['total_sqft']
```

```python
df['price_per_sqft']
```

```python
df['location'] = df['location'].apply(lambda x: x.strip())
```

```python
#extracting the unique value and its count 
loc_count = df['location'].value_counts()
loc_count
```

```python
# extraction those from loc_count whose value is leass than 10
loclessthan10 = loc_count[loc_count<=10]
loclessthan10
```

```python
# replacing values in df[location] which are in loclessthan10 with others
df['location'] = df['location'].apply(lambda x : "others" if x in loclessthan10 else x)
```

**Outlier Detection**

```python
df.describe()
```

```python
(df['total_sqft']/df['bhk']).describe()
```

```python
df = df[(df['total_sqft']/df['bhk']) >= 300]
df.describe()
```

```python
def remove_outlier_sqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf['price_per_sqft'])
        st = np.std(subdf['price_per_sqft'])
        gen_df = subdf[(subdf['price_per_sqft']>(m-st)) & (subdf['price_per_sqft']<=(m+st))]
        df_output = pd.concat([df_output, gen_df], ignore_index = True)
    return df_output
```

```python
df = remove_outlier_sqft(df)
df.describe()
```

```python
def bhk_outlier_remove(df):
    exclude_indices = np.array([])
    for loc, loc_df in df.groupby('location'):
        bhk_stats={}
        for bhk, bhk_df in df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean': np.mean(bhk_df['price_per_sqft']),
                'std' : np.std(bhk_df['price_per_sqft']),
                'count' : bhk_df.shape[0]            
            }
        for bhk, bhk_df in loc_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df['price_per_sqft']<stats['mean']].index.values)
    return df.drop(exclude_indices, axis='index')
```

```python
df = bhk_outlier_remove(df)
```

```python
df = df.drop(columns = ['size','price_per_sqft'])
```

**Machine Learning**

```python
X = df.drop(columns = ['price'])
Y = df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,Y_train, Y_test = train_test_split(X,Y, test_size = 0.2,random_state =42 )
```

```python
print(X_train.shape)
print(X_test.shape)
```

**Linear Regression**

```python
col_transform = make_column_transformer((OneHotEncoder(sparse = False,categories = [df['location'].unique()]), ['location']),remainder = "passthrough")
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(col_transform,scaler,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(Y_test,y_pred_lr)
```

```python
print("Train R2:", pipe.score(X_train, Y_train))
print("Test R2:", pipe.score(X_test, Y_test))
```

**Ridge Regression**

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(col_transform,scaler,ridge)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(Y_test,y_pred_lr)
```

```python
print("Train R2:", pipe.score(X_train, Y_train))
print("Test R2:", pipe.score(X_test, Y_test))
```
